```python
#language translation using hugging face and translation
from transformers import MarianMTModel, MarianTokenizer
# select the model for translation for hugging face
model_name = "Helsinki-NLP/opus-mt-en-hi"
# load the model using model name
model = MarianMTModel.from_pretrained(model_name)
# load the tokenization
tokenizer = MarianTokenizer.from_pretrained(model_name)
#Get the input prompt text from user to translation
prompt = input("Enter the text to translate: ")
# encode the input text into input tokens using tokenizer e.g hi-> 01
input_tokens = tokenizer( prompt, return_tensors='pt', padding=True)
# generate the output tokens using model via data process e.g 40 -> kase
output_tokens = model.generate(**input_tokens)
# decode the output token using tokenizer and generate the final translated text
translated_text = tokenizer.decode(output_tokens[0], skip_special_tokens=True)
#print the translated text
print(translated_text)
```
